```python
# encode characteristics
def encode_characteristic(s, rnd, z0, z1, char):
    if len(z0) <= rnd:
        return
    mask = 0
    b0 = 0
    b1 = 0
    for i in range(32):
        c = char[31-i]
        if c == "0":
            mask |= 1<<i
        elif c == "1":
            b0 |= 1<<i
            b1 |= 1<<i
            mask |= 1<<i
        elif c == "-":
            b0 |= 1<<i
            mask |= 1<<i
        elif c == "+":
            b1 |= 1<<i
            mask |= 1<<i

    nmask = mask ^ 0xFFFFFFFF
    
    # need to encode ^ and !
    s.add((z0[rnd]&mask) == b0)
    s.add((z1[rnd]&mask) == b1)
    
    s.add(z0[rnd]&nmask == z1[rnd]&nmask)
    
    return mask, b0, b1


# encode rules of SHA-1
def encode_sha1(a, w, f, count):
    for i in range(count):
        # encode w
        if i >= 16:
            w.append(RotateLeft(w[-3] ^ w[-8] ^ w[-14] ^ w[-16], 1))
            pass
        
        # encode f
        b = a[-2]
        c = RotateLeft(a[-3], 30)
        d = RotateLeft(a[-4], 30)
        
        if i < 20:
            # if
            f.append(d ^ (b & (c ^ d)))
        elif i < 40 or i >= 60:
            # xor
            f.append(b ^ c ^ d)
        else:
            # maj
            f.append((b&c) ^ (b&d) ^ (c&d))
        
        k = [0x5a827999, 0x6ed9eba1, 0x8f1bbcdc, 0xca62c1d6][i/20]
            
        # encode a
        temp = RotateLeft(a[-1], 5) + f[-1] + RotateLeft(a[-5], 30) + w[i] + k
        a.append(temp)
        
```
